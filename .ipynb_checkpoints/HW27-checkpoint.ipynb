```python
import numpy as np
import pandas as pd
```

```python
csv_df = pd.read_csv(r"mobile.csv")
print(csv_df.info())
```

```python
csv_df["First Name"] = csv_df["Customer Name"].apply(lambda name: name.split(" ")[0])
csv_df.head()
```

```python
def company(num: str)-> str:
    match num:
        case "050":
            return "Pelephone";
        case "051":
            return "Golan";
        case "052":
            return "Cellcom";
        case "053":
            return "Motorola";
        case _:
            return "Unknown Provider";
               
csv_df["Cellular Company"] = csv_df["Phone Number"].apply(lambda num: company(num.split("-")[0]))
csv_df.head()
```

```python
print(csv_df['Cellular Company'].value_counts())
print(list(csv_df['Cellular Company'].unique()))
print("Number of cellproviders: ",csv_df['Cellular Company'].nunique())
```

```python
csv_df['User Type 1'] = csv_df.apply(lambda row: 'Surfer' if row['Internet Minutes'] > row['Talked Minutes'] else "Talker", axis=1)
csv_df['User Type 2'] = np.vectorize(lambda serf, talk:"Surfer" if serf > talk else "Talker")(csv_df['Internet Minutes'], csv_df['Talked Minutes'])
csv_df.head()
```

```python
sort_by_talk = csv_df.sort_values(['Talked Minutes'])
sort_by_talk.head()
```

```python
csv_df['Total Minutes'] = csv_df['Talked Minutes'] + csv_df['Internet Minutes']
sort_by_total = csv_df.sort_values(by='Total Minutes', ascending=False)
sort_by_total.head()
```

```python
sort_by_talk = csv_df.sort_values(['Talked Minutes'], ascending=False)
sort_by_talk.head()
```

```python
sort_by_surf = csv_df.sort_values(['Internet Minutes'])
sort_by_surf.head(3)
```

```python
duplicate_rows = csv_df[csv_df.duplicated()]
print(duplicate_rows)
print()
print(csv_df.drop_duplicates())
```
